```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# Read in the image

img = mpimg.imread("img/test.jpg")
```

```python
plt.imshow(img)
```

```python
print("The image is ",type(img)," with dimensions ",img.shape)
```

```python
# Create copy of the original image for color thresholding

color_select = np.copy(img)
```

```python
# Specify pixel value thresholds for red,green and blue channel

red_threshold = 200
green_threshold = 200
blue_threshold = 200
```

```python
rgb_thresholds = [red_threshold,green_threshold,blue_threshold]
```

```python
thresholds = (color_select[:,:,0]<rgb_thresholds[0]) | (color_select[:,:,1]<rgb_thresholds[1]) | (color_select[:,:,2]<rgb_thresholds[2])
```

```python
color_select[thresholds] = [0,0,0] # Black out the pixels falling below the threshold
```

```python
plt.imshow(color_select)
```
