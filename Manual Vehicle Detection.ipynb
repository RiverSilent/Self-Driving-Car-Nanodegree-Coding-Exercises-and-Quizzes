```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# Read in the image

img = mpimg.imread("img/bbox-example-image.jpg")
plt.imshow(img)
```

```python
bboxes = [((220,500),(370,550)),((820,490),(1150,650))]
```

```python
def draw_boxes(img,bounding_boxes,color=(0,0,255),thick=10):
    for box in bounding_boxes:
        cv2.rectangle(img,box[0],box[1],color,thick)
```

```python
img_copy = np.copy(img)
```

```python
draw_boxes(img_copy,bboxes)
```

```python
plt.imshow(img_copy)
```
