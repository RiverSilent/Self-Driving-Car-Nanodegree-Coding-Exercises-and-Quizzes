```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# Read in the image

img = mpimg.imread("img/stop-sign.jpg")
plt.imshow(img)
```

```python
# Testing out the source points for carrying out the perspective transform

plt.imshow(img)
plt.plot(320,54,".")
plt.plot(324,165,".")
plt.plot(50,217,".")
plt.plot(68,117,".")
```

```python
def warp(img,src,dst):
    img_size = (img.shape[1],img.shape[0])
    M = cv2.getPerspectiveTransform(src,dst)
    Minv = cv2.getPerspectiveTransform(dst,src)
    warped = cv2.warpPerspective(img,M,img_size,flags=cv2.INTER_LINEAR)
    return warped
```

```python
src = np.float32(
    [[320,54],
     [324,165],
     [50,217],
     [68,117]])

dst = np.float32(
    [[320,100],
     [320,220],
     [55,220],
     [55,100]])
```

```python
warped_sign = warp(img,src,dst)
```

```python
plt.imshow(warped_sign)
```
